# Imersão Python Alura - Aula 2 e 3 - Análise de Planilha

#Import

```python
import pandas as pd
import plotly.express as px
```

#Dataframes

```python
df_principal = pd.read_excel('/content/imersaopython.xlsx',sheet_name='Principal')
df_principal.head(5)
```

```python
df_total_acoes = pd.read_excel('/content/imersaopython.xlsx',sheet_name='Total_de_acoes')
df_total_acoes.head(5)
```

```python
df_ticker = pd.read_excel('/content/imersaopython.xlsx',sheet_name='Ticker')
df_ticker.head(5)
```

```python
df_segmento = pd.read_excel('/content/imersaopython.xlsx',sheet_name='gpt')
df_segmento.head(5)
```

# Limpeza do dataframe

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal.head(5)
```

#Colunas renomeadas

```python
df_principal = df_principal.rename(columns={'Ativo':'ativo', 'Data':'data', 'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal.head(5)
```

# Cálculos variação e valor inicial

```python
df_principal['var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct'] + 1)
df_principal.head(5)
```

# Unindo dados de dois dataframes

```python
df_principal=df_principal.merge(df_total_acoes, left_on='ativo',right_on='Código',how='left')
df_principal.head(5)
```

# Coluna repetida removida

```python
df_principal=df_principal.drop(columns=['Código'])
df_principal.head(5)
```

# Cálculo da variação em reais

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtde_teorica'}).copy()
df_principal['var_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['qtde_teorica']
df_principal.head(5)
```

#Formatação números de notação científica para normal.

```python
pd.options.display.float_format='{:,.2f}'.format
df_principal.head(5)
```

```python
df_principal['qtde_teorica'] = df_principal['qtde_teorica'].astype(int)
df_principal.head(5)
```

# Categorias de status

```python
df_principal['resultado'] = df_principal['var_rs'].apply(lambda x: 'subiu' if x > 0 else ('desceu' if x < 0 else 'estavel'))
df_principal
```

# Unindo dataframes

```python
df_principal=df_principal.merge(df_ticker, left_on='ativo',right_on='Ticker',how='left')
df_principal=df_principal.drop(columns=['Ticker'])
df_principal.head(5)
```

```python
df_principal=df_principal.merge(df_segmento, left_on='Nome',right_on='Nome da Empresa',how='left')
df_principal=df_principal.drop(columns=['Nome da Empresa'])
df_principal.head(5)
```

# Renomeando colunas

```python
df_principal = df_principal.rename(columns={'Nome':'nome', 'Segmento':'segmento', 'Idade (em anos)':'idade'}).copy()
df_principal.head(5)
```

# Categorias de idades

```python
df_principal['cat_idade'] = df_principal['idade'].apply(lambda x: 'mais de 100' if x > 100 else ('menos de 50' if x < 50 else 'entre 50 e 100'))
df_principal.head(6)
```

#Análise da variação máxima, mínima e sua média

```python
maior = df_principal['var_rs'].max()
menor = df_principal['var_rs'].min()
media = df_principal['var_rs'].mean()
media_subiu = df_principal[df_principal['resultado'] == 'subiu']['var_rs'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'desceu']['var_rs'].mean()

print(f'maior\tR${maior:,.2f}')
print(f'menor\tR${menor:,.2f}')
print(f'media\tR${media:,.2f}')
print(f'media de quem subiu\tR${media_subiu:,.2f}')
print(f'media de quem desceu\tR${media_desceu:,.2f}')
```

#Tabela nova só com as ações que subiram.

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'subiu']
df_principal_subiu.head(5)
```

# Tabela variações por segmento

```python
df_analise_segmento = df_principal_subiu.groupby('segmento')['var_rs'].sum().reset_index()
df_analise_segmento
```

# Tabela variação por status

```python
df_analise_saldo = df_principal.groupby('resultado')['var_rs'].sum().reset_index()
df_analise_saldo
```

# Gráficos

```python
fig = px.bar(df_analise_saldo, x='resultado', y='var_rs', text='var_rs', title='Variação em reais')
fig.update_traces(texttemplate='R$%{y:,.2f}', textposition='outside')
fig.show()
```

```python
fig = px.pie(df_analise_segmento, values='var_rs', names='segmento', title='Variação por segmento')
fig.show()
```

```python
df_analise_idade = df_principal_subiu.groupby('cat_idade')['var_rs'].sum().reset_index()
df_analise_idade
```

Fiz o gráfico relacionado à tabela anterior.

```python
fig = px.bar(df_analise_idade, x='cat_idade', y='var_rs', text='var_rs', title='Variação em reais por idade')
fig.update_traces(texttemplate='R$%{y:,.2f}', textposition='outside')
fig.show()
```
